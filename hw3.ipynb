__Problem 3:__ (15 points) Write a function that takes in a string as input and returns a list of the unique letters used in the string. Punctuation should not be counted as a unique letter. For our purposes you may consider the following characters as punctuation: `.,;:?!-`  Your list should return all letters as lower case and sorted. So for example if the input is ' `Happy days are here again!`' your function should return the list `['a''d','e','g','h','i','n','p','r','s','y']`

```python
string_input = input('Type something.')
#convert to lower case
lower_input = string_input.lower()
#remove spaces
lower_input = lower_input.replace(' ','')
#remove repetitions 
unique_input = set(lower_input)
#convert set to list 
list_input = list(unique_input)
#sort 
list_input.sort()
#remove punctuations
punctuation = ['.',',',';',':','?','!','-']
string_without_punctuation = []
for item in list_input:
    if item not in punctuation: 
        string_without_punctuation.append(item)
print(string_without_punctuation)
```

* length_n(file_name, n):  returns an list of all words of length *n*.
* starts_with(file_name, n, first_letter):  returns a list of all words of length *n* beginning with the letter *first_letter*
* contains_letter(file_name, n, included):  returns an list of words of length *n* containing the letter *included* but not beginning with it.



```python
def length_n(file_name,n):
     # your code here
    open_file = open(file_name,'r')
    word = open_file.readline()
    word = word.rstrip('\n')
    while word != '':
        if len(word) == n:
            print(word)
        word = open_file.readline()
        word = word.rstrip('\n')
    open_file.close()

#try running the code
length_n(file_name,4)
```

```python
def starts_with(file_name, n, first_letter):
    # your code here
    open_file = open(file_name,'r')
    word = open_file.readline()
    word = word.rstrip('\n')
    while word != '':
        if len(word) == n and word[0] == first_letter:
            print(word)
        word = open_file.readline()
        word = word.rstrip('\n')
    open_file.close()

#try running the code
starts_with('dictionary.txt',4,'i')
```
